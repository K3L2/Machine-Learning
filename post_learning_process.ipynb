```python
import ultralytics
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os
```

```python
# .pt 파일 path 지정

best_path = 'best (240725_epoch200).pt'
```

```python
model = YOLO(best_path)
```

```python
# img 들을 담고있는 dir_path 지정해기기

img_dir_path = 'save_img'
img_list = os.listdir(img_dir_path)
res_list = []
```

```python
for i in range(len(img_list)):
    img_path = os.path.join(img_dir_path, img_list[i]).replace('\\','/')
    result = model(img_path)[0]
    res_list.append(result.plot())
```

```python
len(res_list)
```

```python

# figsize와 rows, cols 설정하면 됩니다.
fig = plt.figure(figsize=(12,8))
rows = 8
cols = 5
i = 1

for img in res_list:
    ax = fig.add_subplot(rows, cols, i)
    ax.imshow(img)
    ax.set_xticks([]), ax.set_yticks([])
    i += 1
```

```python
# figsize 값을 크게 설정하여 서브플롯 크기를 키움
fig = plt.figure(figsize=(24, 16))  # 예: (24, 16)으로 변경

rows = 8
cols = 5
i = 1

for img in res_list:
    ax = fig.add_subplot(rows, cols, i)
    ax.imshow(img)
    ax.set_xticks([]), ax.set_yticks([])
    i += 1

# 서브플롯 간의 간격을 조정
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# 플롯을 화면에 표시
plt.show()
```
